# **Practical - 2 (Line Graph with Python)**

# list of common types of line charts:

## 1.Simple Line Chart

- Displays a single line representing one data series.

## 2.Multiple Line Chart

- Shows multiple lines in one chart to compare different data series.

## 3.Stacked Line Chart

- Lines are stacked on top of each other to show the cumulative values of multiple data series.

## 4.Stepped Line Chart

- The lines follow a step-like progression, ideal for visualizing discrete changes.

## 5.Dashed or Styled Line Chart

- Lines with different styles (dashed, dotted, etc.) for enhanced visualization.

## 6.Line Chart with Markers

- Combines lines with markers to emphasize specific data points.

## 7.Logarithmic Line Chart

- Uses a logarithmic scale for the y-axis, suitable for exponential growth trends.

## 8.Line Chart with Error Bars

- Includes error bars to represent variability or uncertainty in data.

## 9.Interactive Line Chart

- Allows interactivity like zooming, panning, or hovering to show details (commonly created with tools like Plotly or Bokeh).

## 10.Time-Series Line Chart

- Tracks changes over time, often used for trends analysis.

## 11.Dual-Axis Line Chart

- Features two vertical axes to plot two different data series with distinct units.

## 12.Polar Line Chart

- Plots data in polar coordinates, useful for cyclic data (e.g., wind directions).

## 13.3D Line Chart

- Adds a third dimension to visualize relationships between three variables.

## 14.Area Line Chart

- Similar to a line chart but with the area under the line filled to emphasize magnitude.

## 15.Spline Line Chart

- Uses smooth curves instead of straight lines for connecting data points.

## 16.Threshold Line Chart

- Highlights a specific threshold (e.g., a target value or limit) across the chart.


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import LogLocator
import plotly.graph_objects as go
```

```python
# Generate simple data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # Quadratic
y2 = [5, 4, 3, 2, 1]    # Decreasing linear
y3 = [2, 2, 2, 2, 2]    # Constant
```

```python
# 1. Simple Line Chart
plt.figure(figsize=(6, 4))
plt.plot(x, y1, label='y = x^2')
plt.title('Simple Line Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Line Graph
plt.figure(figsize=(6, 4))
plt.plot(x, y1, label='y1', color='Red')
plt.title('Line Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 2. Multiple Line Chart
import matplotlib.pyplot as plt
# Data for plots
x = [1, 2, 3, 4, 5]
x1 = [2, 5, 7, 9, 10]
y1 = [1, 4, 9, 16, 25]
y2 = [2, 6, 8, 22, 30]
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x1,y2)
plt.title('Multiple Line Chart - Your name')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 3. Stacked Line Chart
plt.figure(figsize=(6, 4))
plt.stackplot(x, y1, y2, y3, labels=['y = x^2', 'y = decreasing', 'y = constant'], alpha=0.6)
plt.title('Stacked Line Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 4. Stepped Line Chart
plt.figure(figsize=(6, 4))
plt.step(x, y1, label='y = x^2', where='mid')
plt.title('Stepped Line Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5. Dashed or Styled Line Chart
plt.figure(figsize=(6, 4))
plt.plot(x, y1, linestyle='--', label='Dashed y = x^2')
plt.plot(x, y2, linestyle=':', label='Dotted y = decreasing')
plt.title('Dashed or Styled Line Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 6. Line Chart with Markers
plt.figure(figsize=(6, 4))
plt.plot(x, y1, marker='o', label='y = x^2 with markers')
plt.title('Line Chart with Markers')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 7. Logarithmic Line Chart
plt.figure(figsize=(6, 4))
plt.plot(x, y1, label='y = x^2')
plt.yscale('log')
plt.gca().yaxis.set_major_locator(LogLocator(base=10.0))
plt.title('Logarithmic Line Chart')
plt.xlabel('x')
plt.ylabel('y (log scale)')
plt.legend()
plt.grid(True, which='both')
plt.show()
```

```python
# 8. Line Chart with Error Bars
errors = [0.5, 0.4, 0.3, 0.2, 0.1]
plt.figure(figsize=(6, 4))
plt.errorbar(x, y1, yerr=errors, fmt='-o', label='y = x^2 with error bars')
plt.title('Line Chart with Error Bars')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 9. Interactive Line Chart (using Plotly)
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines+markers', name='y = x^2'))
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='y = decreasing'))
fig.update_layout(title='Interactive Line Chart', xaxis_title='x', yaxis_title='y')
fig.show()
```

```python
# 10. Time-Series Line Chart
# Simulating time-series data
time = ['2025-01-01', '2025-01-02', '2025-01-03', '2025-01-04', '2025-01-05']
y_time = [3, 5, 2, 6, 4]
plt.figure(figsize=(6, 4))
plt.plot(time, y_time, marker='o', label='Time-Series Data')
plt.title('Time-Series Line Chart')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 11. Dual-Axis Line Chart
fig, ax1 = plt.subplots(figsize=(6, 4))
ax2 = ax1.twinx()
ax1.plot(x, y1, 'g-', label='y = x^2')
ax2.plot(x, y2, 'b--', label='y = decreasing')
ax1.set_xlabel('x')
ax1.set_ylabel('y1 (green)', color='g')
ax2.set_ylabel('y2 (blue)', color='b')
plt.title('Dual-Axis Line Chart')
plt.grid(True)
plt.show()
```

```python
# 12. Polar Line Chart
angles = np.linspace(0, 2 * np.pi, len(x))
r = y1
plt.figure(figsize=(6, 6))
ax = plt.subplot(111, polar=True)
ax.plot(angles, r, label='Polar y = x^2')
ax.set_title('Polar Line Chart')
plt.legend()
plt.show()
```

```python
# 13. 3D Line Chart
from mpl_toolkits.mplot3d import Axes3D
z = [0, 1, 2, 3, 4]
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y1, z, label='3D Line')
ax.set_title('3D Line Chart')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.legend()
plt.show()
```

```python
# 14. Area Line Chart
plt.figure(figsize=(6, 4))
plt.fill_between(x, y1, color="skyblue", alpha=0.4)
plt.plot(x, y1, color="Slateblue", alpha=0.6, linewidth=2, label='Area under y = x^2')
plt.title('Area Line Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 15. Spline Line Chart
from scipy.interpolate import make_interp_spline
x_new = np.linspace(min(x), max(x), 300)
y1_smooth = make_interp_spline(x, y1)(x_new)
plt.figure(figsize=(6, 4))
plt.plot(x_new, y1_smooth, label='Spline y = x^2')
plt.title('Spline Line Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 16. Threshold Line Chart
threshold = 15
plt.figure(figsize=(6, 4))
plt.plot(x, y1, label='y = x^2')
plt.axhline(y=threshold, color='r', linestyle='--', label='Threshold')
plt.title('Threshold Line Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# **Practical - 9 (Import data from kaggle and visualize the data)**

# **Practical - 9 With Iris DataSet**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd  # Import pandas for data manipulation
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target # Y- Axis
feature_names = iris.feature_names
target_names = iris.target_names

# Create a DataFrame with species and sepal length
df = pd.DataFrame({'Species': target_names[target], 'Sepal Length': data[:, 0]})

# Bar chart of sepal length by species using the DataFrame
plt.figure(figsize=(8, 6))
sns.barplot(x='Species', y='Sepal Length', hue='Species', data=df)  # Use the DataFrame
plt.title('Bar Chart of Sepal Length by Species = Name: ')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()


# Pie chart of species distribution
plt.figure(figsize=(6, 6))
# Calculate the frequency of each species
species_counts = {species: list(target).count(i) for i, species in enumerate(target_names)}
# Get the species names and their corresponding counts
species = list(species_counts.keys())
counts = list(species_counts.values())
# Create the pie chart using the species names and counts
plt.pie(counts, labels=species, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Species Distribution')
plt.show()


# Scatter plot of sepal length vs. sepal width
plt.figure(figsize=(8, 6))
for i, species in enumerate(target_names):
    plt.scatter(data[target == i, 0], data[target == i, 1], label=species)
plt.title('Scatter Plot of Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()


# Histogram of sepal length
plt.figure(figsize=(8, 6))
plt.hist(data[:, 0], bins=10)
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
